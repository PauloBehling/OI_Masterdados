## Valor do documento, valor glosa e valor líquido

O valor do documento representa o valor total da despesa para a qual o parlamentar solicitou a restituição. O valor glosa representa a parcela de retenção de pagamentos para despesas irregulares ou que extrapolam o limite de gastos. O valor líquido representa o montante efetivamente restituído ao parlamentar.

Portanto, iremos considerar somente o valor líquido das despesas, uma vez que ele representa o valor real dos gastos da união com despesas parlamentares.

```python
pt_despesas = dados.loc[(dados['sgPartido'] == 'PT') & (dados['vlrLiquido'].notnull())][variaveis]
pt_despesas.head(3)
```

```python
pt_media_despesas = pt_despesas['vlrLiquido'].mean()
todas_media_despesas = todas_despesas['vlrLiquido'].mean()

print('Quantidade de transações do PT: ', pt_despesas.shape[0])
print('Valor médio das transações do PT: R$ ', round(pt_media_despesas, 2))
print('Quantidade de transações de todos: ', todas_despesas.shape[0])
print('Valor médio das transações de todos: R$ ', round(todas_media_despesas, 2))
```

```python
pt_acima_media_despesas = pt_despesas.loc[pt_despesas['vlrLiquido'] > pt_media_despesas]
todas_acima_media_despesas = todas_despesas.loc[todas_despesas['vlrLiquido'] > todas_media_despesas]

print('Quantidade de transações do PT acima do valor médio: ', 
    pt_acima_media_despesas.shape[0])
print('Valor médio das transações do PT acima do valor médio: R$ ', round(pt_acima_media_despesas['vlrLiquido'].mean(), 2))
print('Quantidade de transações de todos acima do valor médio: ', todas_acima_media_despesas.shape[0])
print('Valor médio das transações de todos acima do valor médio: R$ ', round(todas_acima_media_despesas['vlrLiquido'].mean(), 2))
```
